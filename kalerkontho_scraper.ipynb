```

source = requests.get('https://www.kalerkantho.com/print-edition/2021/10/08').content
soup = BeautifulSoup(source, 'html.parser')
csv_file = open('cms_scrape.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'summary', 'video_link'])
for article in soup.find_all('div', class_='col-sm-8 cat_home'):
  #print(article)
  headline=article.text
  print(headline)
csv_file.close()
```
